```python
Sales = float(input("Enter sales ($): "))
Tax = float(input("Enter tax rate (%): "))
print("Sales: $" ,Sales)
COGS = Sales * .4
print("Cost of Goods Sold: $" ,COGS)
Profit = Sales - COGS
print("Gross Profit: $", Profit)
Tax = Profit * (Tax / 100)
print("Tax: $", Tax)
Income = Profit - Tax
print("Net Income: $", Income)
```

```python
first_number = float(input("Enter first number: "))
second_number = float(input("Enter second number: "))
operator = input("Enter operator: ")

if operator == "+":
  print(first_number + second_number)
elif operator == "-":
  print(first_number - second_number)
elif operator == "*":
  print(first_number * second_number)
elif operator == "/":
  print(first_number / second_number)
elif operator == "**":
  print(first_number ** second_number)
elif operator == "%":
  print(first_number % second_number)
elif operator == "//":
  print(first_number // second_number)
else:
  print("Invalid Input")
```

```python
import math

ticketPrice = 10
initAttendees = 20
fixedCost = 200

print("Ad$\tProfit")

for ad in range(0, 201, 25):
  attendees = initAttendees + 2 * round(math.sqrt(ad))
  cost = ad + fixedCost
  revenue = attendees * ticketPrice
  profit = revenue - cost
  print(f"{float(ad)}\t{float(profit)}")
```

```python
def reverse(s):
  if len(s) < 2:
    return s
  else:
    return s[-1] + reverse(s[1:-1]) + s[0] #swaps first and last characters

reverse('asjduh1')
```

```python
def loan_payment(principal, annual_rate, months):
  assert (isinstance(principal, (int,float)) and
          isinstance(annual_rate, (int,float)) and
          isinstance(months, int))
  if (principal <= 0) or (annual_rate < 0) or (months <= 0):
    raise ValueError("Invalid Input")

  try:
    r = annual_rate / 12
    if r == 0:
      pmt = principal / months
    else:
      pmt = r * principal / (1 - (1 + r) ** (-months))
  except Exception as e:
    print(f"Error: {e}")
```

```python
class Greeting:
  def __init__ (self, first, last):
    self.first = first
    self.last = last

  def display(self, formal=False):
    if formal:
      print(f"{self.last}, {self.first}")
    else:
      print(self.first)

  def __str__(self):
    return "Instance of class 'Greeting!'"

name = Greeting("Jane", "Doe")
print(name)
name.display()
name.display(formal=True)

```

```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.number = number
    self.__courses = []

  def get_name(self):
    return self.__name

  def get_courses(self):
    return self.__courses

  def set_name(self, newName):
    self.__name = newName

  def __len__(self):
    return len(self.__courses)

  def enroll(self, newCourse):
    if newCourse not in self.__courses:
      self.__courses.append(newCourse)
      print(f"Enrolled in {newCourse}")
    else:
      print(f"Student {self.__name} is already enrolled in {newCourse}")
```

```python
s = Student("Jane Doe", 12345)
print(f"Student name: {s.get_name()}")
print(f"Show currently enrolled courses: {s.get_courses()}")
s.enroll('BTE 320')
print(f"Show currently enrolled courses: {s.get_courses()}")
s.enroll('BTE 320')
print(f"Total number of courses enrolled: {len(s)}")
```
