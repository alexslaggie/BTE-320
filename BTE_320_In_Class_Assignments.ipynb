```python
Sales = float(input("Enter sales ($): "))
Tax = float(input("Enter tax rate (%): "))
print("Sales: $" ,Sales)
COGS = Sales * .4
print("Cost of Goods Sold: $" ,COGS)
Profit = Sales - COGS
print("Gross Profit: $", Profit)
Tax = Profit * (Tax / 100)
print("Tax: $", Tax)
Income = Profit - Tax
print("Net Income: $", Income)
```

```python
first_number = float(input("Enter first number: "))
second_number = float(input("Enter second number: "))
operator = input("Enter operator: ")

if operator == "+":
  print(first_number + second_number)
elif operator == "-":
  print(first_number - second_number)
elif operator == "*":
  print(first_number * second_number)
elif operator == "/":
  print(first_number / second_number)
elif operator == "**":
  print(first_number ** second_number)
elif operator == "%":
  print(first_number % second_number)
elif operator == "//":
  print(first_number // second_number)
else:
  print("Invalid Input")
```

```python
base = 5
extra = 2.5

print("Hour\tFee")
for hour in range(1,9):
  fee = base+extra*hour
  if fee < 10.0:
    fee = 10.0
  elif fee > 20.0:
    fee = 20.0
  print(f"{hour}\t{fee}")
```

```python
import math

ticketPrice = 10
initAttendees = 20
fixedCost = 200

print("Ad$\tProfit")

for ad in range(0, 201, 25):
  attendees = initAttendees + 2 * round(math.sqrt(ad))
  cost = ad + fixedCost
  revenue = attendees * ticketPrice
  profit = revenue - cost
  print(f"{float(ad)}\t{float(profit)}")
```

```python
base = 5
extra = 2.5
hour = 1

print("Hour\tFee")
while hour < 9:
  fee = base+extra*hour
  if fee < 10.0:
    fee = 10.0
  elif fee > 20.0:
    fee = 20.0
  print(f"{hour}\t{fee}")
  hour += 1
```
