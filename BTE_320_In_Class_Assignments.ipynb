```python
def calcFee(hours):
    base = 5
    extra = 2.5

    fee = base + extra * hours

    if fee < 10.0:
        fee = 10.0
    elif fee > 20.0:
        fee = 20.0

    return fee


def main():
    # print table
    print("Hour\tFee")
    for hour in range(1, 9):
        print(f"{hour}\t{calcFee(hour)}")

    # user input
    hours = float(input("\nEnter number of parking hours: "))
    fee = calcFee(hours)
    print(f"Parking fee: ${fee}")


main()
```

```python
def loan_payment(principal, annual_rate, months):
  assert (isinstance(principal, (int,float)) and
          isinstance(annual_rate, (int,float)) and
          isinstance(months, int))
  if (principal <= 0) or (annual_rate < 0) or (months <= 0):
    raise ValueError("Invalid Input")

  try:
    r = annual_rate / 12
    if r == 0:
      pmt = principal / months
    else:
      pmt = r * principal / (1 - (1 + r) ** (-months))
  except Exception as e:
    print(f"Error: {e}")
```

```python
def getValues(n):
    values = []
    for i in range(n):
        while True:
            try:
                val = float(input(f"Enter value {i+1}: "))
                if val < 0:
                    raise ValueError
                values.append(val)
                break
            except:
                print("Invalid input. Enter a non-negative number.")
    return values

def calcAverage(alist):
    blist = alist[3:]
    return sum(blist) / len(blist)

n = int(input("How many values? "))
values = getValues(n)
avg = calcAverage(values)

print(f"Average (excluding first 3 values): {avg}")
```

```python
class Greeting:
  def __init__ (self, first, last):
    self.first = first
    self.last = last

  def display(self, formal=False):
    if formal:
      print(f"{self.last}, {self.first}")
    else:
      print(self.first)

  def __str__(self):
    return "Instance of class 'Greeting!'"

name = Greeting("Jane", "Doe")
print(name)
name.display()
name.display(formal=True)

```

```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.number = number
    self.__courses = []

  def get_name(self):
    return self.__name

  def get_courses(self):
    return self.__courses

  def set_name(self, newName):
    self.__name = newName

  def __len__(self):
    return len(self.__courses)

  def enroll(self, newCourse):
    if newCourse not in self.__courses:
      self.__courses.append(newCourse)
      print(f"Enrolled in {newCourse}")
    else:
      print(f"Student {self.__name} is already enrolled in {newCourse}")
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    dx = (self.x - other.x) ** 2
    dy = (self.y - other.y) ** 2
    return (dx + dy) ** 0.5

class Rectangle:
    def __init__(self, p1, p2, p3, p4):
        self.p1 = p1
        self.p2 = p2
        self.p3 = p3
        self.p4 = p4

        self.width = self.calc_width()
        self.height = self.calc_height()

    def calc_width(self):
        x_values = [self.p1.x, self.p2.x, self.p3.x, self.p4.x]
        return max(x_values) - min(x_values)

    def calc_height(self):
        y_values = [self.p1.y, self.p2.y, self.p3.y, self.p4.y]
        return max(y_values) - min(y_values)

    def calc_area(self):
        return self.width * self.height

p1 = Point(0, 0)
p2 = Point(10, 0)
p3 = Point(10, 5)
p4 = Point(0, 5)

rect = Rectangle(p1, p2, p3, p4)

print(rect.width)
print(rect.height)
print(rect.calc_area())
```

```python
class Employee:
  def __init__(self, name, wage, hours):
    self.name = name
    self.__hourly_wage = wage
    self.hours = hours

  def get_wage(self):
    return self.__hourly_wage

  def set_wage(self, newWage):
    assert newWage > 0, "Wage values cannot be negative or zero."
    self.__hourly_wage = newWage

  def calculate_pay(self):
    return self.__hourly_wage * self.hours

  def __str__(self):
    return f"Employee: {self.name}\nHourly wages: {self.__hourly_wage}"


class Manager(Employee):
  def __init__(self, name, wage, hours, bonus):
    super().__init__(name, wage, hours)
    self.bonus = bonus # assume it is given as float 0 -> 1 (e.g., 20% bonus is 0.2)

  def calculate_pay(self):
    return (1 + self.bonus) * self.get_wage() * self.hours # (1 + self.bonus) * super().calculate_pay()

employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)
print(employee)
print(manager)
print()
print(f"Employee {employee.name} weekly salary is ${employee.calculate_pay()}.")
print(f"Manager {manager.name} weekly salary is ${manager.calculate_pay()} that includes a {manager.bonus * 100}% bonus.")
```
